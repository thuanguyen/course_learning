```python
from graphviz import Digraph

def de_bruijn_ize(s, k):
    nodes, edges = set(), []
    for i in range(len(s) - k + 1):
        p, q = s[i:i+k-1], s[i+1:i+k]
        nodes |= {p, q}
        edges.append((p, q))
    return nodes, edges


def draw_de_bruijn(s, k, filename="debruijn_graph", fmt="png"):
    # Build graph
    nodes, edges = de_bruijn_ize(s, k)
    dot = Digraph(format=fmt)

    # Add nodes and edges
    for n in nodes:
        dot.node(n)
    for u, v in edges:
        dot.edge(u, v)

    # Save graph to a file
    outpath = dot.render(filename, cleanup=True)
    print(f"Saved graph to: {outpath}")

    return dot


# Example usage
g = draw_de_bruijn("ACGCGTCG", 3, filename="my_debruijn_graph", fmt="png")
g   # display in Jupyter / Vertex
```

```python
if __name__ == "__main__":
    # Tiny example
    s = "ACGCGTCG"
    k = 3
    nodes, edges = de_bruijn_ize(s, k)
    print("Nodes:", nodes)
    print("Edges:")
    for u, v in edges:
        print(f"{u} -> {v}")
```
