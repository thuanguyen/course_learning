```python
from graphviz import Digraph

def de_bruijn_ize(s, k):
    nodes, edges = set(), []
    for i in range(len(s) - k + 1):
        p, q = s[i:i + k - 1], s[i + 1:i + k]
        nodes |= {p, q}
        edges.append((p, q))
    return nodes, edges


def draw_de_bruijn(s, k):
    nodes, edges = de_bruijn_ize(s, k)
    dot = Digraph()
    for n in nodes:
        dot.node(n)
    for u, v in edges:
        dot.edge(u, v)
    return dot
```

```python
if __name__ == "__main__":
    # Tiny example
    s = "ACGCGTCG"
    k = 3
    nodes, edges = de_bruijn_ize(s, k)
    print("Nodes:", nodes)
    print("Edges:")
    for u, v in edges:
        print(f"{u} -> {v}")
```

```python
# Example
draw_de_bruijn("ACGCGTCG", 3)
```

```python
# Example
draw_de_bruijn("a_long_long_long_time", 5)
```
