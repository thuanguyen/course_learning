```python
# Q2

import random
def create_dna(n, alphabet='acgt'):
    return ''.join([random.choice(alphabet) for i in range(n)])
dna = create_dna(4)
print(dna)
```

```python
#Q4
import time

def count4(dna, base): # the build-in function is the fastest one
    return dna.count(base)

t01 = time.clock()
t02 = time.time()
print('count4:', count4(dna = '', base = 'C'))
print(time.clock() - t01)
print(time.time() - t02)
```
